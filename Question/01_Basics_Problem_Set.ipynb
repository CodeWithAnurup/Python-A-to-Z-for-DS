22. **Basic Calculator (Menu Driven)**

Create a menu with `+`, `-`, `*`, `/` options. Input two numbers and perform the selected operation.

```python
a = int(input("a = "))
b = int(input("b = "))
menu = input()

match menu:
    case "+":
        print(a + b) 
    case "-":
        print(a - b)
    case "*":
        print(a * b)
    case "/":
        print(a/b)
    case _:
        print("You used wrong symbol")
    
```

26. **Unit Converter Tool**

Convert kilograms to pounds, centimeters to inches, etc., based on user selection.

```python
value = int(input("Enter your choice:\n1. Kilograms <-> Pounds\n2. Centimeters <-> Inches\nChoice: "))

match value:
    case 1:
        choice = int(input("\n1. Kilograms to Pounds\n2. Pounds to Kilograms\nChoice: "))

        match choice:
            case 1:
                kg = float(input("Enter weight in kilograms: "))
                print(f"{kg} kg = {kg * 2.20462:.2f} pounds")
            case 2:
                pound = float(input("Enter weight in pounds: "))
                print(f"{pound} pounds = {pound / 2.20462:.2f} kg")
            case _:
                print("Invalid choice.")

    case 2:
        choice = int(input("\n1. Centimeters to Inches\n2. Inches to Centimeters\nChoice: "))

        match choice:
            case 1:
                cm = float(input("Enter length in centimeters: "))
                print(f"{cm} cm = {cm / 2.54:.2f} inches")
            case 2:
                inch = float(input("Enter length in inches: "))
                print(f"{inch} inches = {inch * 2.54:.2f} cm")
            case _:
                print("Invalid choice.")

    case _:
        print("Invalid main menu choice.")
```

31. **Student Grade Calculator**

Input marks of 5 subjects, calculate average, and determine grade (`A`, `B`, etc.).

```python
n = int(input("How many subject student has- "))
print("The grade-A = avr more than 95, grade-B = avr more than 70, grade-C = avr more than 60")
sub = []
for i in range(0,n):
    mark = int(input(f"The marks of subject{i+1}"))
    sub.append(mark)

avr = sum(sub)/n
print(f"The average of {n} subject is = {avr}")

if 90 < avr:
    print("Grade-A")
elif 70 < avr:
    print("Grade-B")
elif 60 < avr:
    print("Grade-C")
else:
    print("Grade-D")
```

### 42. **Return sum of two distinct postion of a list by given target**

```python
def two_sum(num, target):
    for i in range(len(num)):
        for j in range(len(num)):
            if i != j:      
                if num[i] + num[j] == target:
                    return i, j

num = [3, 7, 3, 15] 
target = 6
two_sum(num, target)
```
