8. **Check for Even Equality**

Check if the square of a number is equal to its double.

```python
def double(num):
    return 2 * num 

def square(num):
    return num*num

num = int(input())
if double(num) == square(num):
    print ("equal")
else:
    print(f"{double(num)} not equal to {square(num)}")
```
