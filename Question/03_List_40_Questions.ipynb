```python
numbers = [12, 13, 1, 7, 8 ,23, 7 ,9, 7]
number = [12, 13, 1, 7, 8 , 23, 7 ,9, 7]

def compare(numbers, number):
    if len(number) == len(numbers): 
        if sorted(number) == sorted(numbers):
            return True
        else:
            return False
    else:
        return False
                

compare(number, numbers)
```

## 35. 🧠 Filter Based on Condition
Filter students whose names start with a vowel from a list of names.

```python
fruits = ['Apple', 'pineapple', 'Banana', 'coConut', 'strawbery']

vowel = ('a', 'e', 'i', 'o', 'u')
fruit = []
new_fruit = []
for j in fruits:
    fruit.append(j.lower())
    
for p in fruit:
    if p[0] in vowel:
        new_fruit.append(p)
        print(new_fruit)
```

## 36. 📊 Student Gradebook App
Store student names and marks in a list. Print name, mark, and grade (A/B/C) for each.

```python
mark = ["Shantanu" , 75]

if mark[1] > 90:
    print("A")
elif mark[1] > 70:
    print("B")
elif mark[1] < 45:
    print("C")
```

```python

student_names = [
    "Aarav", "Isha", "Rohan", "Sneha", "Karan", "Pooja", "Aditi", "Ankit", "Riya", "Suresh",
    "Meena", "Siddharth", "Aman", "Kavya", "Divya", "Raj", "Mohit", "Neha", "Rahul", "Priya",
    "Aditya", "Simran", "Tushar", "Preeti", "Yash", "Bhavna", "Nikhil", "Tanvi", "Deepak", "Komal",
    "Gaurav", "Shruti", "Nitin", "Pallavi", "Manish", "Payal", "Vikram", "Reena", "Alok", "Sonam"
]

marks = [
    45, 67, 22, 89, 30, 76, 49, 15, 90, 28,
    60, 55, 12, 33, 31, 25, 47, 80, 29, 38,
    50, 41, 20, 19, 65, 34, 27, 32, 58, 73,
    24, 26, 35, 44, 48, 17, 21, 39, 28, 36
]

# Create dictionary
student_dict = dict(zip(student_names, marks))

print("Name\t\tMarks\tGrade")
print("-" * 30)

for student_name, mark in student_dict.items():

    if mark >= 90:
        grade = "A"
    elif mark >= 80:
        grade = "B"
    elif mark >= 70:
        grade = "C"
    elif mark >= 60:
        grade = "D"
    elif mark >= 45:
        grade = "E"
    else:
        grade = "F"

    print(f"{student_name:<10}\t{mark}\t{grade}")
```

## 37. 📅 Daily Planner
Create a weekly planner. Let the user add, remove, and view tasks using list operations.

```python
plan = []
print("1 for add, 2 for remove, 3 for view, 4 for break")

def Add(x):
    plan.append(x)
def Remove(x):
    plan.remove(x)
def View():
    print(plan)

n = int(input("How many command you wanna run- "))

print("choose a opstion in between 1 2 3 and 4- ")

for i in range(0,n):
    ops = int(input())
    if ops == 1:
        x = int(input("Appending- "))
        Add(x)
    elif ops == 2:
        r = int(input("Removing a number- "))
        Remove(r)
    elif ops == 3:
        print("This is your list-")
        View()
    else:
        if ops == 4:
            print("Out of program.")
            break

    
    print(plan)
```
