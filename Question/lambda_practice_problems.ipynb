## Section A: Lambda Basics

**1. Square a Number**

Write a lambda that takes a number and returns its square.

```python
x = 2

square = lambda x: x*x
print(square(x))
```

**2. Larger of Two Numbers**

```python
a = 6
b = 1

larger = lambda a,b: max(a,b)
print(larger(a,b))
```

**8. Long Words**

From a list of words, filter out only the ones with length greater than 5.

```python
fruits = ['apple', 'pineapple', 'banana', 'coconut', 'strawbery']

long_words = list(filter(lambda x: len(x) > 5, fruits))
print(long_words)
```

```python
fruits = ['apple', 'pineapple', 'banana', 'coconut', 'strawbery']

long_words = list(filter(lambda x: len(x)>5, fruits))
print(long_words)
```

**9. Remove Vowels**

From a list of single characters, filter out only the consonants (remove vowels).

```python
alphabet = ['a', 'e', 'i', 'o', 'u' , 'x' , 'z' , 'c' , 'm' , 'r']
vowels = list(filter(lambda x : x  in ['a', 'e', 'i', 'o' , 'u'], alphabet))
print(vowels)
```

**10. Passing Students**

Given a list of tuples `[("Aman", 78), ("Riya", 32), ("Karan", 45)]`, filter out students who scored 40 or above.

```python
students = [("Aman", 78), ("Riya", 32), ("Karan", 45)]

passnumber = filter(lambda x : x[1]>=40, students)
names =list(map(lambda x : x[0], passnumber))
print(names)
```

**11. Multiples of 3**

From a list of numbers, filter out only the multiples of 3.

```python
number = [3, 4, 6, 7, 9, 10, 12, 13, 15, 16, 18, 20, 21]

three = list(filter(lambda x : x%3 == 0, number))
print(three)
```

## Section D: Lambda + `sorted()` (using `key=`)

**17. Sort Tuples by Second Element**

Sort `[("Aman", 78), ("Riya", 32), ("Karan", 90)]` by marks (the second element), highest first.

```python
std = [("Aman", 78), ("Riya", 32), ("Karan", 90)]

sorting = sorted(std, key = lambda x : x[1], reverse=True)
print(sorting)
```
